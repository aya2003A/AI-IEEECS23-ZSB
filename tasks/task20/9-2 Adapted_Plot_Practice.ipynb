```python
# prerequisite package imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline
```

In this workspace, you will work with the fuel economy dataset from the previous lesson on bivariate plots.

```python
fuel_econ = pd.read_csv("/Users/Hp/Downloads/fuel-econ.csv")
fuel_econ.head()
```

**Task 1**: Plot the city ('city') vs. highway ('highway') fuel efficiencies (both in mpg) for each vehicle class ('VClass'). Don't forget that vehicle class is an ordinal variable with levels {Minicompact Cars, Subcompact Cars, Compact Cars, Midsize Cars, Large Cars}.

```python
vehicle_class_order = ['Minicompact Cars', 'Subcompact Cars', 'Compact Cars', 'Midsize Cars', 'Large Cars']
plt.figure(figsize=(10, 6))
sb.scatterplot(x='city', y='highway', hue='VClass', hue_order=vehicle_class_order, data=fuel_econ)

plt.xlabel('City Fuel Efficiency')
plt.ylabel('Highway Fuel Efficiency')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sb.violinplot(x='VClass', y='city', order=vehicle_class_order, data=fuel_econ, jitter=True, dodge=True, alpha=0.7, size=6, label='City')
sb.violinplot(x='VClass', y='highway', order=vehicle_class_order, data=fuel_econ, jitter=True, dodge=True, alpha=0.7, size=6, color='orange', label='Highway')
plt.xlabel('Vehicle Class')
plt.ylabel('Fuel Efficiency')
plt.legend(title='Fuel Efficiency')
plt.show()
```

**Task 2**: Plot the relationship between engine size ('displ', in liters), vehicle class, and fuel type ('fuelType'). For the lattermost feature, focus only on Premium Gasoline and Regular Gasoline cars. What kind of relationships can you spot in this plot?

```python
fuel_efficiency_data_subset = fuel_econ[fuel_econ['fuelType'].isin(['Premium Gasoline', 'Regular Gasoline'])]

plt.figure(figsize=(12, 8))
sb.scatterplot(x='VClass', y='displ', hue='fuelType', data=fuel_efficiency_data_subset, size=4)

plt.xlabel('Engine Size')
plt.ylabel('Vehicle Class')

plt.show()


```
